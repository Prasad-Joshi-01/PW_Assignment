```python
X= df.iloc[:, :-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.decomposition import PCA

pca = PCA()
X_pca = pca.fit_transform(X)
```

```python
import numpy as np
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

# Find the number of principal components that explain at least 95% of the variance
n_components = np.argmax(cumulative_variance_ratio >= 0.95) + 1
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Scatter Plot')
plt.show()
```

```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore")

kmeans = KMeans(n_clusters=3)
kmeans.fit(X_pca)
cluster_labels = kmeans.labels_
```

```python
cluster_labels
```

```python
Summary : PCA helps in reducing the dimensionality of the dataset and identifying the most important features that explain the variance in the data. 
The scatter plot visualizes the data in the reduced-dimensional space, with each data point colored based on its class label.
K-Means clustering is then performed on the PCA-transformed data to group similar data points together.
The resulting cluster labels can be used to understand patterns and similarities in the data.
```
