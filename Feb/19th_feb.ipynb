```python
Q1. What is Flask Framework? What are the advantages of Flask Framework?
Ans:
Flask is a micro web framework written in Python that is used for building web applications.
Flask is designed to be lightweight, modular, and easy to use, and it is often used for building small to medium-sized web applications.

Here are some of the advantages of Flask framework:

Lightweight and flexible: Flask is a lightweight framework that provides just the basic functionality needed for building web applications, which makes it very flexible and customizable. 
Developers can easily extend Flask with their own libraries and tools as needed.

Easy to learn and use: Flask is known for its simplicity and ease of use. 
It has a simple syntax and requires only a few lines of code to get a basic web application up and running.
It also has a large community of developers who create and maintain plugins and extensions that can be used to add more functionality to a Flask application.

Built-in development server: Flask comes with a built-in development server that makes it easy to test and debug applications locally.
```

```python
Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.
Ans:
In Flask, the url_for function is used for URL building. 
This function generates a URL based on the name of the endpoint, which is associated with a view function.

The url_for function takes the endpoint name as its first argument, and optionally, additional keyword arguments representing the variable parts of the URL.
```

```python
from flask import Flask,url_for
from flask import request

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('index'))
```
