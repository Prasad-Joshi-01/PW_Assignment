```python
Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
Ans:
It represents errors that occur when a lookup or indexing operation fails, such as when an item is not found in a dictionary or a list.
```

```python
KeyError: This exception is raised when attempting to access a key in a dictionary that does not exist
```

```python
my_dict = {'a': 1, 'b': 2, 'c': 3}
try:
    value = my_dict['d']
except KeyError as e:
    print("Error:", e)
```

```python
IndexError: This exception is raised when attempting to access an index in a list that is out of range.
```

```python
my_list = [1, 2, 3]
try:
    value = my_list[3]
except IndexError as e:
    print("Error:", e)
```

```python
Q5. Explain ImportError. What is ModuleNotFoundError?
Ans:
ImportError: It is a built-in exception class in Python that is raised when an imported module, package, or object cannot be found or loaded.
```

```python
try:
    import my_module
except ImportError as e:
    print("An ImportError occurred:", e)
```

```python
ModuleNotFoundError: It is a built-in exception class in Python that is raised when an imported module or package cannot be found.
```

```python
Q6. List down some best practices for exception handling in python.
```

```python
1.Catch specific exceptions: Catch only the exceptions that you expect to occur and that you can handle in a meaningful way.
Avoid catching general exceptions like Exception or BaseException, as they can mask unexpected errors and make debugging more difficult.

2.Use the try-except-else block: Use the try-except-else block to handle exceptions and separate error handling from the main code logic. 
The else block allows you to run code that should only be executed if the try block completes without raising an exception.

3.Use the finally block for cleanup: Use the finally block to run code that should be executed regardless of whether an exception was raised or not. 
This block is typically used for cleanup tasks, such as closing files or database connections.

4.Log exceptions: Use Pythons logging module to log exceptions and other important information about your programs execution.
This can be helpful for debugging and monitoring your application.

5.Raise meaningful exceptions: When defining custom exceptions, make sure they provide enough information to understand the problem that occurred.
Include a clear error message and any relevant data that can help diagnose the issue.

6.Dont ignore exceptions: Never ignore exceptions or fail to handle them properly.
This can lead to unexpected behavior and make debugging more difficult. Always handle exceptions in a meaningful way, even if it means logging an error message and exiting the program.

7.Avoid bare except clauses: Avoid using bare except clauses that catch all exceptions without specifying the exception type. 
This can make debugging more difficult and lead to unexpected behavior.
```
