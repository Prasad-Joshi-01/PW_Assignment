```python
Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
Ans:
In Bokeh, glyphs are visual shapes or markers that can be added to a plot to represent data points. 
Examples of glyphs include circles, squares, triangles, and lines.
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
p = figure(title='Bokeh plot with glyphs', x_axis_label='X-axis', y_axis_label='Y-axis')
p.circle(x, y)
p.square(x, y)
output_notebook()
show(p)
```

```python
Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
```

```python
from bokeh.plotting import figure, show
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
#customized title
p = figure(title="Headline example")
p.line(x, y, legend_label="Temp.", line_width=2)
p.title_location = "left"
p.title.text = "Changing headline text example"
#customized axes
p.xaxis.axis_label = "Temp"
p.xaxis.axis_line_color = "red"
p.yaxis.axis_label = "Pressure"
p.yaxis.major_label_orientation = "vertical"
#customized Legend
p.legend.location = "top_left"
p.legend.border_line_color = "navy"

show(p)
```

```python
Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?
Ans:
A Bokeh server is a Python process that runs a Bokeh application, allowing us to create interactive plots that can be updated in real time. 
With a Bokeh server, we can create web-based applications that allow users to interact with your data and 
update the plot dynamically, without needing to reload the entire page. 
Bokeh provides a 'curdoc()' function that represents the current document, which we can use to add Bokeh objects to the document. 
we can also define callback functions that respond to user input or other events, and update the data or properties of the Bokeh objects in the document. 
The Bokeh server provides a web socket connection between the clientand the server, allowing for real-time updates to the plot based on user input.
```

```python
Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?
Ans:
To embed a Bokeh plot into a web page or dashboard using Flask or Django, we can use the 'components()' function provided by Bokeh. 
This function generates the necessary HTML and JavaScript code to embed the Bokeh plot into a web page.

Here is the basic process for embedding a Bokeh plot in a Flask or Django app:
Create a Bokeh plot in your Flask or Django application.
Use the components() function to generate the necessary HTML and JavaScript code for the plot.
Pass the generated script and div components to your Flask or Django template.
Render the template with the plot components.

In Flask, you would typically define a route function that generates the Bokeh plot and passes the plot components to a template,
 which is then rendered by the Flask application.

In Django, you would typically define a view function that generates the Bokeh plot and passes the plot components to a template,
 which is then rendered by the Django application.
```
