```python
Matplotlib is a Python library used for data visualization.
It provides a wide range of tools for creating different types of plots, charts, and graphs from data.
Matplotlib is widely used in scientific computing, data analysis, and machine learning.

Here are some common types of plots that can be created using the Pyplot module of Matplotlib:
1.Line plot - used to visualize the trend or relationship between two variables over time
2.Scatter plot - used to visualize the relationship between two variables as points on a two-dimensional plane
3.Bar plot - used to compare the values of different categories using rectangular bars
4.Histogram - used to visualize the distribution of a single variable by dividing it into bins
5.Pie chart - used to show the proportion of each category as a slice of a circular pie.
```

```python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.
```

```python
A scatter plot is a type of plot used to display the relationship between two variables.
It is often used to explore the correlation or association between two continuous variables.
```

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title")
plt.show()
```

```python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

```python
The subplot() function in Matplotlib is used to create multiple plots in a single figure. 
It provides a way to display different subplots within a larger plot area, which can be useful for comparing different sets of data or 
displaying different views of the same data.
```

```python
import matplotlib.pyplot as plt
```

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

axs[0, 0].set_title('Line Plot 1')
axs[0, 1].set_title('Line Plot 2')
axs[1, 0].set_title('Line Plot 3')
axs[1, 1].set_title('Line Plot 4')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y')
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('Y')
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Y')
axs[1, 1].set_xlabel('X')
axs[1, 1].set_ylabel('Y')

plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.show()
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
Ans:
A bar plot is a type of chart that displays data with rectangular bars. 
The height or length of each bar is proportional to the value it represents. 
Bar plots are often used to show comparisons between different categories or groups of data.
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
import matplotlib.pyplot as plt
plt.bar(company,profit)
plt.xlabel("company")
plt.ylabel("profit")
plt.title("profit w.r.t company")
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
Ans:
A box plot, also known as a box and whisker plot, is a type of chart that displays the distribution of a dataset.
Box plots are used to visualize the spread, skewness, and outliers of a dataset.
They are particularly useful for comparing the distribution of different datasets or variables, and for detecting outliers or unusual values.
```

```python
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

a = [box1, box2]
fig, ax = plt.subplots()
ax.boxplot(a)

ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Data Values')

plt.show()
```
